### Birthday Paradox 
I read something like this:
- __In probability theory, the birthday problem asks for the probability that, in a set of n randomly chosen people, at least two will share a birthday. The birthday paradox is that, counterintuitively, the probability of a shared birthday exceeds 50% in a group of only 23 people.__
----

Formulation for this: 
- probability = 1 - [(364 x 363 x 362 x ... x (365 - n)) / 365]  
- n = # of people in the room

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def shared_birthday_prob(n):
    not_shared = 1
    for i in range(365, (365 -n ), -1):
        not_shared *= (i/365)
    return round(1 - not_shared, 2)
```

```python
shared_birthday_prob(8)
```

```python
probs = []
number_of_people = []

for i in range(1, 100):
    number_of_people.append(i)
    probs.append(shared_birthday_prob(i))

df = pd.DataFrame({
    'probs': probs,
    'number_of_people': number_of_people
})
```

```python
plt.figure(figsize=(20, 15))
plt.scatter(x = df['number_of_people'], y = df['probs'])

for i in range(len(df)//3):
    plt.annotate(df.loc[i, 'probs'], (df.iloc[i, 1], df.iloc[i, 0]+0.05))
```
